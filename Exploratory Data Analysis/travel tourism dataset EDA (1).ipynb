```python
revenue_df['Average Revenue Growth'] = revenue_df.loc[revenue_df['Category'] == 'REVENUE GROWTH in percent', '2017':'2023'].mean(axis=1)
revenue_df['Combined ARPU'] = revenue_df.loc[revenue_df['Category'] == 'ARPU in thousand USD (US$)', '2017':'2023'].sum(axis=1)
revenue_df['Revenue Std Dev'] = revenue_df.loc[revenue_df['Category'] == 'REVENUE in billion USD (US$)', '2017':'2023'].std(axis=1)

print(revenue_df)
```
